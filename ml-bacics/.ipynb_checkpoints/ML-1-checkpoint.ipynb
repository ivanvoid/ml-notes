### (ML 1.2) What is supervised learning?

Classes of ML problems  
Supervised vs Unsupervised   
  
**Supervised:** Given data  $(x_1, y_1),...,(x_n, y_n)$,  
where  
$x_i$ - data point  
$y_i$ - class / value  
Task:  
Choose a function $f(x) = y$ 

Types of supervised learning:
- Classification (when $y_i \in $ { finite set } )
- Regression (when $y_i \in {\rm I\!R}^d$)

```python
# Loading libs and setup random seed
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(42)

# Generating data
x1 = np.linspace(0,1,10)
x2 = np.random.normal(x1)
y = np.zeros(len(x1))
y[:len(y)//2] = 1

new_point = [0.4, 1.0]

# Plot the data
plt.plot(x1[y == 0],x2[y == 0], 'go', label='class 0')
plt.plot(x1[y == 1],x2[y == 1], 'ro', label='class 1')

plt.plot(new_point[0],new_point[1], 'bx', label='new point')

plt.title('Classification problem')
plt.xlabel('x1')
plt.ylabel('x2')
plt.legend()
plt.show()
```

```python
# Generating data
x = np.linspace(0,10,10) 
x += np.random.normal(x)
y = np.random.normal(x)

plt.plot(x,y, 'ko')

plt.title('Regression problem')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

### [(ML 1.3) What is unsupervised learning?](https://www.youtube.com/watch?v=lANt56UOaSk&list=PLD0F06AA0D2E8FFBA&index=3)
**Unsupervised:** Given: $(x_i,...x_n)$  
where $x_i \in {\rm I\!R}^d$  
Task:  
find patterns in the data  
- Clustering 
- Density estimation
- Demensionality reduction

```python
from sklearn.datasets import make_blobs

# generate 2d classification dataset
x, _ = make_blobs(n_samples=100, centers=3, n_features=2)

plt.plot(x[:,0], x[:,1], 'ko')

plt.title('Unsupervised problem')
plt.xlabel('x1')
plt.ylabel('x2')
plt.show()
```

**Density estimation.**  
you assume that the data comes from some probability distribution with some density, and you wanna estimate that density.

```python
np.random.seed(42)
x = np.random.normal(0, 3, 30)
plt.plot(x, np.zeros(30), 'ko')

count, bins, ignored = plt.hist(x, 30, density=True)
plt.plot(bins, 1/(3 * np.sqrt(2 * np.pi)) *
               np.exp( - (bins - 0)**2 / (2 * 3**2) ),
         linewidth=2, color='r')

plt.show()
```
